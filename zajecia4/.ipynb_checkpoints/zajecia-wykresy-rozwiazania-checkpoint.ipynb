## Scatterplot

```python
plt.scatter(heart_data['age'], heart_data['chol'])
plt.xlabel("Wiek")
plt.ylabel("Cholesterol")
plt.show()
plt.close()
```

## Histogram

```python
import matplotlib.pyplot as plt
plt.hist(heart_data['age'])
plt.ylabel("Wiek")
plt.xlabel("Liczba osob")
plt.title("Rozklad wieku w zbiorze danych")
plt.show()
plt.close()
```

## Wiele wykresów 

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.hist(heart_data['age'], bins=20)
plt.title("Wiek")

plt.subplot(2, 2, 2)
plt.hist(heart_data['chol'])
plt.title("Cholesterol")

plt.subplot(2, 2, 3)
plt.hist(heart_data['trestbps'])
plt.title("Cisnienie")

plt.subplot(2, 2, 4)
plt.hist(heart_data["thalach"])
plt.title("Maks_bicie_serca")

plt.show()
plt.close()
```

Stwórz figure gdzie znajdą się 4 wykresy: wiek vs cholesterol, wiek vs cisnienie, cholesterol vs cisnienie, cholesterol vs maks_bicie_serca

Dodaj do każdego wykresu xlabel, ylabel, title i zmień liczbę bins

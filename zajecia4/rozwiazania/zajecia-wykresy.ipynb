## Wiele wykresów 

````python
import numpy as np
import matplotlib.pyplot as plt

# rozmiar końcowego zbioru wykresów
plt.figure(figsize=(15,10))

# pierwszy wykres
plt.subplot(2, 2, 1)
plt.hist(heart_data['age'], bins=20)
plt.title("Wiek")

# drugi wykres
plt.subplot(2, 2, 2)
plt.hist(heart_data['chol'])
plt.title("Cholesterol")

# trzeci wykres
plt.subplot(2, 2, 3)
plt.hist(heart_data['trestbps'])
plt.title("Cisnienie")

# czwarty wykres
plt.subplot(2, 2, 4)
plt.hist(heart_data["thalach"])
plt.title("Maks bicie serca")

plt.show()
plt.close()
````

Stwórz figure gdzie znajdą się 4 wykresy: wiek vs cholesterol, wiek vs cisnienie, cholesterol vs cisnienie, cholesterol vs maks_bicie_serca

````python
import numpy as np
import matplotlib.pyplot as plt

# rozmiar końcowego zbioru wykresów
plt.figure(figsize=(15,10))

# pierwszy wykres
# wiek vs cholesterol
plt.subplot(2, 2, 1)
plt.scatter(heart_data['age'], heart_data['chol'])
plt.xlabel("Wiek")
plt.ylabel("Cholesterol")
plt.title("wiek vs cholesterol")

# drugi wykres
# wiek vs cośnienie
plt.subplot(2, 2, 2)
plt.scatter(heart_data['age'], heart_data['trestbps'])
plt.xlabel("Wiek")
plt.ylabel("Cisnienie")
plt.title("wiek vs cisnienie")

# trzeci wykres
# cholesterol vs ciśnienie
plt.subplot(2, 2, 3)
plt.scatter(heart_data['chol'], heart_data['trestbps'])
plt.xlabel("Cholesterol")
plt.ylabel("Cisnienie")
plt.title("cholesterol vs cisnienie")

# czwarty wykres
# cholesterol vs maks_bicie_serca
plt.subplot(2, 2, 4)
plt.scatter(heart_data['chol'], heart_data['thalach'])
plt.xlabel("Cholesterol")
plt.ylabel("Maks bicie serca")
plt.title("cholesterol vs maks bicie serca")

plt.show()
plt.close()
````

Dodaj do każdego wykresu xlabel, ylabel, title

## Zapisywanie wykresów
Wykresy zapisujemy używając fukcji
```
plt.savefig("ścieżka z nową nazwą pliku")
```

Musimy jednak zrobić to przed `plt.close()`, bo ta funkcja usuwa z pamięci stworzony wykres!

````python
plt.scatter(heart_data['age'].values, heart_data['chol'].values)
plt.xlabel("Wiek")
plt.ylabel("Cholesterol")
plt.show()
plt.savefig('wiek_cholesterol.png')
plt.close()
````

## Wybierz część tabeli gdzie osoby mają < 50 lat

````python
mlodzi = heart_data[heart_data['age'] < 50]
````

Jaka jest średnia wieku w tej grupie?

````python
mlodzi.mean()
````

Narysuj histogramy (jak wyżej) dla tej grupy.

````python
import numpy as np
import matplotlib.pyplot as plt

# rozmiar końcowego zbioru wykresów
plt.figure(figsize=(15,10))

# pierwszy wykres
plt.subplot(2, 2, 1)
plt.hist(mlodzi['age'], bins=20)
plt.title("Wiek")

# drugi wykres
plt.subplot(2, 2, 2)
plt.hist(mlodzi['chol'])
plt.title("Cholesterol")

# trzeci wykres
plt.subplot(2, 2, 3)
plt.hist(mlodzi['trestbps'])
plt.title("Cisnienie")

# czwarty wykres
plt.subplot(2, 2, 4)
plt.hist(mlodzi["thalach"])
plt.title("Maks bicie serca")

plt.show()
plt.close()
````

Norma cholesterolu we krwi to do 175 mg/dl. Ile osób mieści się w normie? Ile osób < 50 lat mieści się w normie?

````python
w_normie = heart_data[heart_data["chol"] < 175]
print(w_normie.shape)
print(heart_data.shape)

w_normie_mlodzi = mlodzi[mlodzi["chol"] < 175]
print(w_normie_mlodzi.shape)
print(mlodzi.shape)
````
